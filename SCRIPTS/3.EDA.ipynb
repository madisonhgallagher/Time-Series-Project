```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
CO_data = pd.read_csv('/content/LA_CO_2024.csv')

# Ensure the date column is in datetime format
CO_data['Date'] = pd.to_datetime(CO_data['Date'])

# Plot the ozone levels over time
plt.figure(figsize=(12, 8))
plt.plot(CO_data['Date'], CO_data['Daily Max 8-hour CO Concentration'], label='CO Levels', color='blue')

# Add labels and title
plt.title('CO Levels Over Time in 2024', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('CO Levels', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
ozone_data = pd.read_csv('/content/LA_Ozone_2024.csv')

# Ensure the date column is in datetime format
ozone_data['Date'] = pd.to_datetime(ozone_data['Date'])

# Filter the data to only include rows where 'Source' column is 'AQS'
filtered_data = ozone_data[ozone_data['Source'] == 'AQS']

# Plot the ozone levels over time
plt.figure(figsize=(12, 8))
plt.plot(filtered_data['Date'], filtered_data['Daily Max 8-hour Ozone Concentration'], label='Ozone Levels (Source: AQS)', color='blue')

# Add labels and title
plt.title('Ozone Levels Over Time in 2024 (Filtered by Source: AQS)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Ozone Levels', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
ozone_data = pd.read_csv('/LA_Ozone_2024.csv')

# Ensure the date column is in datetime format
ozone_data['Date'] = pd.to_datetime(ozone_data['Date'])

# Calculate the daily average ozone levels
daily_avg_ozone = ozone_data.groupby(ozone_data['Date'].dt.date)['Daily Max 8-hour Ozone Concentration'].mean()

# Convert the index back to datetime for plotting
daily_avg_ozone.index = pd.to_datetime(daily_avg_ozone.index)

# Plot the daily average ozone levels over time
plt.figure(figsize=(12, 8))
plt.plot(daily_avg_ozone.index, daily_avg_ozone.values, label='Daily Average Ozone Levels', color='blue')

# Add labels and title
plt.title('Daily Average Ozone Levels Over Time (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Ozone Levels', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
pm2_data = pd.read_csv('/LA_PM2.5_2024.csv')

# Ensure the date column is in datetime format
pm2_data['Date'] = pd.to_datetime(pm2_data['Date'])

# Calculate the daily average PM2.5 levels
daily_avg_pm2 = pm2_data.groupby(pm2_data['Date'].dt.date)['Daily Mean PM2.5 Concentration'].mean()

# Convert the index back to datetime for plotting
daily_avg_pm2.index = pd.to_datetime(daily_avg_pm2.index)

# Plot the daily average PM2.5 levels over time
plt.figure(figsize=(12, 8))
plt.plot(daily_avg_pm2.index, daily_avg_pm2.values, label='Daily Average PM2.5 Levels', color='green')

# Add labels and title
plt.title('Daily Average Small Particle Matter Levels Over Time (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average PM2.5 Levels', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert each Series into a DataFrame
df_ozone = daily_avg_ozone.to_frame(name='ozone')
df_co = daily_avg_co.to_frame(name='co')
df_pm2 = daily_avg_pm2.to_frame(name='pm25')
df_pm10 = daily_avg_pm10.to_frame(name='pm10')

# Merge all DataFrames on their datetime indices
merged_data = pd.concat([df_ozone, df_co, df_pm2, df_pm10], axis=1)

# Display a preview of the merged DataFrame
print(merged_data.head())

# Plot the merged data
plt.figure(figsize=(12, 8))
plt.plot(merged_data.index, merged_data['ozone'], label='Ozone', color='blue')
plt.plot(merged_data.index, merged_data['co'], label='Carbon Monoxide (CO)', color='red')
plt.plot(merged_data.index, merged_data['pm25'], label='PM2.5', color='orange')
plt.plot(merged_data.index, merged_data['pm10'], label='PM10', color='green')

# Add labels, title, and legend
plt.title('Daily Averages of Air Quality Measurements (2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Concentration', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```
